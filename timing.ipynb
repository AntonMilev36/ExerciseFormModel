```python
import import_ipynb

from feature_healpers import find_descent_frames, find_ascent_frames, find_bottom_pause
```

```python
def descent_time(video_df, fps):
    descent = find_descent_frames(video_df)
    return len(descent) / fps


def ascent_time(video_df, fps):
    ascent = find_ascent_frames(video_df)
    return len(ascent) / fps


def bottom_pause(video_df, fps):
    pause_frames = find_bottom_pause(video_df)
    return pause_frames / fps
```
