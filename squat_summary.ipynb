```python
import import_ipynb

from range_of_motion import (maximum_squat_depth, lowest_hip_position, 
                             minimum_knee_angle, minimum_hip_angle, 
                             maximum_torso_lean, maximum_knee_forward, 
                             maximum_hip_shift)
from timing import descent_time, ascent_time, bottom_pause
from motion import average_descent_speed, average_ascent_speed, maximum_acceleration, maximum_jerk
from stability import hip_x_std, knee_x_std, shoulder_x_std
from coordination import hip_knee_correlation, shoulder_hip_correlation
from normalized_ratios import depth_body_height, knee_travel_depth, torso_lean_depth
from label import label
```

```python
def estimate_fps(video_df):
    return len(video_df) / (video_df["frame"].max() - video_df["frame"].min() + 1)
```

```python
def summarize_video(video_df):
    fps = estimate_fps(video_df)

    return {
        "maximum_squat_depth": maximum_squat_depth(video_df),
        "lowest_hip_position": lowest_hip_position(video_df),
        "minimum_knee_angle": minimum_knee_angle(video_df),
        "minimum_hip_angle": minimum_hip_angle(video_df),
        "maximum_torso_lean": maximum_torso_lean(video_df),
        "maximum_knee_forward": maximum_knee_forward(video_df),
        "maximum_hip_shift": maximum_hip_shift(video_df),

        "descent_time": descent_time(video_df, fps),
        "ascent_time": ascent_time(video_df, fps),
        "bottom_pause": bottom_pause(video_df, fps),

        "average_descent_speed": average_descent_speed(video_df),
        "average_ascent_speed": average_ascent_speed(video_df),
        "maximum_acceleration": maximum_acceleration(video_df),
        "maximum_jerk": maximum_jerk(video_df),

        "hip_x_std": hip_x_std(video_df),
        "knee_x_std": knee_x_std(video_df),
        "shoulder_x_std": shoulder_x_std(video_df),

        "hip_knee_correlation": hip_knee_correlation(video_df),
        "shoulder_hip_correlation": shoulder_hip_correlation(video_df),

        "depth_body_height": depth_body_height(video_df),
        "knee_travel_depth": knee_travel_depth(video_df),
        "torso_lean_depth": torso_lean_depth(video_df),

        "label": label(video_df)
    }
```
