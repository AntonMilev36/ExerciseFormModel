```python
import import_ipynb

from range_of_motion import maximum_squat_depth, maximum_knee_forward, maximum_torso_lean 
```

```python
def depth_body_height(video_df):
    depth = maximum_squat_depth(video_df)
    body_height = video_df["body_height"].mean()

    return depth / body_height if body_height != 0 else 0


def knee_travel_depth(video_df):
    depth = maximum_squat_depth(video_df)
    knee = maximum_knee_forward(video_df)

    return knee / depth if depth != 0 else 0


def torso_lean_depth(video_df):
    depth = maximum_squat_depth(video_df)
    lean = maximum_torso_lean(video_df)

    return lean / depth if depth != 0 else 0
```
