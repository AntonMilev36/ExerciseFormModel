```python
import import_ipynb

from feature_healpers import find_descent_frames, find_ascent_frames
```

```python
def average_descent_speed(video_df):
    descent = find_descent_frames(video_df)
    return descent["hip_velocity"].mean()


def average_ascent_speed(video_df):
    ascent = find_ascent_frames(video_df)
    return ascent["hip_velocity"].mean()


def maximum_acceleration(video_df):
    return video_df["hip_acceleration"].abs().max()


def maximum_jerk(video_df):
    return video_df["hip_jerk"].abs().max()
```
