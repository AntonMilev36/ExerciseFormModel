```python
def maximum_squat_depth(video_df):
    """
    Difference between highest and lowest hip position.
    """
    return video_df["hip_height"].max() - video_df["hip_height"].min()


def lowest_hip_position(video_df):
    """
    Lowest hip position reached.
    """
    return video_df["hip_height"].max()


def minimum_knee_angle(video_df):
    return video_df["knee_angle"].min()


def minimum_hip_angle(video_df):
    return video_df["hip_angle"].min()


def maximum_torso_lean(video_df):
    return video_df["torso_angle"].max()


def maximum_knee_forward(video_df):
    return video_df["knee_forward"].max()


def maximum_hip_shift(video_df):
    return video_df["hip_shift"].max()
```
